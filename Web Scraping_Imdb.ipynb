```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import pandas as pd
```

```python
base_site = input('Please insert the link for the imdb page')
response = requests.get(base_site)
print(response.status_code)
html = response.content
soup = BeautifulSoup(html, "lxml")
div_notes= soup.find_all("div", {"class": "filmo-category-section"})[0]
div_links = [div.find('a') for div in div_notes]
links= []
for i in range(0,len(div_links)):
    if i%2==0:
        continue
    else:
        link= div_links[i]
        links.append(link)
link_urls = [urljoin(base_site, l.get('href')) for l in links]
titles=[]
years=[]
scores=[]
directors=[]
for link in link_urls:
    note_resp = requests.get(link)
    if note_resp.status_code == 200:
                    note_html = note_resp.content    
                    note_soup = BeautifulSoup(note_html, 'lxml')
                    note_soup.find("h1", {"class": "TitleHeader__TitleText-sc-1wu6n3d-0 dxSWFG"})
                    if note_soup.find("h1", {"class": "TitleHeader__TitleText-sc-1wu6n3d-0 dxSWFG"}) != None:
                        is_series= 'Series' in note_soup.find("h1", {"class": "TitleHeader__TitleText-sc-1wu6n3d-0 dxSWFG"})
                        if is_series == False:   ## We just concentrate on movies, not Tv series
                            title= note_soup.find("h1", {"class": "TitleHeader__TitleText-sc-1wu6n3d-0 dxSWFG"}).text
                            if note_soup.find('span', {"class": "TitleBlockMetaData__ListItemText-sc-12ein40-2 jedhex"}) != None:
                                year= int(note_soup.find('span', {"class": "TitleBlockMetaData__ListItemText-sc-12ein40-2 jedhex"}).text[:4])
                            else:
                                 continue 
                            score= note_soup.find("div", {"class": "RatingBar__RatingContainer-sc-85l9wd-0 hNqCJh Hero__HideableRatingBar-kvkd64-12 hBqmiS"}).text[11:14]
                            director= note_soup.find("a", {"class": "ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link"}).text
                            titles.append(title)
                            years.append(year)
                            scores.append(score)
                            directors.append(director)
                        else:
                              continue
                    else:
                          continue
    else:
            continue
movie= pd.DataFrame([titles, years, scores, directors]).T
movie.rename(columns={0:"Title",1: "Year",2: "Score", 3: "Director"}, inplace= True)
movie
```
